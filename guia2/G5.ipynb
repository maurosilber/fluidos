## 1) Rotación y helicidad cinética

Considere un fluído incompresible en un sistema rotante con velocidad angular $\pmb{\Omega}$,

\begin{aligned}
\frac{\partial \textbf{u}}{\partial t} + (\textbf{u} \cdot \nabla) \textbf{u} = - \nabla \left( \frac{p}{\rho} \right) + 2 \textbf{u} \times \pmb{\Omega} + \nu \nabla^2 \textbf{u} \tag{1}
\end{aligned}

> a) Muestre que para condiciones periódicas, la helicidad cinética,
\begin{aligned}
H \equiv \int_V \textbf{u} \cdot \pmb{\omega} \, dV
\end{aligned}
> satisface una ecuación de balance y se conserva cuando $\nu = 0$
>
> b) Para el caso en que $\nu \neq 0$, muestre que la tasa de disipación de helicidad cinética está controlada por,
\begin{aligned}
S \equiv \int_V \pmb{\omega} \times (\nabla \times \pmb{\omega}) \, dV
\end{aligned}
>
> c) La helicidad cinética es una forma de cuantificar si un flujo posee (o no) simetría de reflexión. ¿Qué ocurre con el signo de $\textit{H}$ ante reflexiones?

### a) Ecuación de balance para la helicidad

Para poder construir la ecuación de balance de la helicidad
querríamos partir de la primer ecuación de $\textbf{u}$
y construir términos que contengan $\textbf{u} \cdot \pmb{\omega}$.
Lo que haremos será entonces aplicar el rotor a esta ecuación
y de esa forma tener una segunda ecuación para la dinámica de $\pmb{\omega}$.
Luego haremos la suma de $\pmb{\omega} \cdot (1) + \textbf{u} \cdot (2)$.

---

Aplicando el rotor,
obtenenemos la siguiente ecuación para $\pmb{\omega}$:

$$ \frac{\partial \pmb{\omega}}{\partial t}
- (\pmb{\omega} \nabla)\textbf{u}
+ (\textbf{u} \nabla) \pmb{\omega}
= 2 (\pmb{\Omega} \nabla) \textbf{u}
+ \nu \nabla^2 \pmb{\omega}
\tag{2}
$$

---

Realizando la suma de $\pmb{\omega} \cdot (1) + \textbf{u} \cdot (2)$,
obtenemos los siguientes términos:

$$ \underbrace{\pmb{\omega} \frac{\partial \textbf{u}}{\partial t} 
+ \textbf{u} \frac{\partial \pmb{\omega}}{\partial t}}_{(1)} 
+ \underbrace{\pmb{\omega} (\textbf{u} \cdot \nabla) \textbf{u} 
+ \textbf{u} (\textbf{u} \cdot \nabla) \pmb{\omega}}_{(2)}
- \underbrace{ \textbf{u} (\pmb{\omega} \cdot \nabla)\textbf{u}}_{(3)}
= - \underbrace{\pmb{\omega} \nabla \left( \frac{p}{\rho} \right)}_{(4)} 
+ \underbrace{2 \pmb{\omega} (\textbf{u} \times \pmb{\Omega}) 
+ 2 \textbf{u} (\pmb{\Omega} \cdot \nabla) \textbf{u}}_{(5)} 
+ \underbrace{\nu \, \pmb{\omega} \nabla^2 \textbf{u} 
+ \nu \, \textbf{u} \nabla^2 \pmb{\omega}}_{(6)}  \tag{3}
$$

Desarrollamos cada término de forma que surja le helicidad
y que el resto de los términos queden escritos como la divergencia de algo.

### Desarrollo de los términos

**Término 1**

$$ \pmb{\omega} \frac{\partial \textbf{u}}{\partial t} 
+ \textbf{u} \frac{\partial \pmb{\omega}}{\partial t} 
= \frac{\partial (\textbf{u} \cdot \pmb{\omega})}{\partial t}
$$

---

**Término 2**

$$ \pmb{\omega} (\textbf{u} \cdot \nabla) \textbf{u} 
+ \textbf{u} (\textbf{u} \cdot \nabla) \pmb{\omega} 
= \nabla[\textbf{u}(\textbf{u} \cdot \pmb{\omega})]
$$
---

**Término 3**

$$ \textbf{u} (\pmb{\omega} \cdot \nabla)\textbf{u} 
= \nabla \left( \pmb{\omega} \frac{u^2}{2} \right)
$$
---

**Término 4**

$$ \pmb{\omega} \nabla \left( \frac{p}{\rho} \right) 
= \nabla \left(\pmb{\omega} \frac{p}{\rho} \right)
$$

---

**Término 5**

$$
2 \pmb{\omega} (\textbf{u} \times \pmb{\Omega}) 
+ 2 \textbf{u} (\pmb{\Omega} \cdot \nabla) \textbf{u} 
= \nabla[\textbf{u} (\Omega \cdot \textbf{u})]
$$

---

**Término 6**

Este término no necesitamos desarrollarlo todavía.
Podemos dejarlo como está
y ya en el punto **b** lo trabajamos
para hacer aparecer la magnitud $\textit{S}$.

Reemplazamos los términos calculados en la ecuación $(3)$
e integramos respecto del volumen.
Esto nos queda entonces como:

\begin{aligned}
\frac{\partial }{\partial t} \underbrace{\int_V \textbf{u} \cdot \pmb{\omega} \, dV}_{\equiv H} 
& = \underbrace{\int_V \{ -\nabla[\textbf{u}(\textbf{u} \cdot \pmb{\omega})] 
+ \nabla \left( \pmb{\omega} \frac{u^2}{2} \right) 
- \nabla \left( \pmb{\omega} \frac{p}{\rho} 
+ \nabla[\textbf{u}(\textbf{u} \cdot \pmb{\Omega})] \right) \} \, dV }
_{= 0, \text{ Utilizando Gauss y condiciones de contorno periódicas}} 
+  \nu \int_V \pmb{\omega} \nabla^2 \textbf{u} \, dV 
+ \nu \int_V \textbf{u} \nabla^2 \pmb{\omega} \, dV  \\
\frac{\partial H}{\partial t} 
& = \nu \int_V \left(  \pmb{\omega} \nabla^2 \textbf{u}
+ \textbf{u} \nabla^2 \pmb{\omega} \right) \, dV
\end{aligned}

De esta manera,
si $\nu = 0$ la helicidad del sistema se conserva.

### b) Disipación de la helicidad

Dada la identidad de Green:

\begin{aligned}
\textbf{P} (\nabla \times \nabla \times \textbf{Q})
- \textbf{Q} (\nabla \times \nabla \times \textbf{P}) =
\nabla (\textbf{Q} \times (\nabla \times \textbf{P}) 
- \textbf{P} \times (\nabla \times \textbf{Q}))
\quad \quad (4)
\end{aligned}

Reemplazando $\textbf{P} = \pmb{\omega}$ y $\textbf{Q} = \textbf{u}$ :

\begin{aligned}
\pmb{\omega}(\nabla \times \nabla \times \textbf{u}) 
- \textbf{u}(\nabla \times \nabla \times \pmb{\omega})
&=
\nabla[\textbf{u} \times (\nabla \times \pmb{\omega})
- \underbrace{\pmb{\omega} \times (\nabla \times \textbf{u})}_{\pmb{\omega} \times \pmb{\omega} = 0 }] \\\\
\pmb{\omega}(\nabla \times \pmb{\omega})
&=
\underbrace{\nabla[\textbf{u} \times (\nabla \times \pmb{\omega})]}_{(1)} 
+ \underbrace{\textbf{u}(\nabla \times \nabla \times \pmb{\omega})}_{(2)}
\end{aligned}

El término de la izquierda es el que queremos obtener.
Por tanto,
tenemos que trabajar los términos de la derecha
y vincularlos con los términos laplacianos de la ecuación $\textit{(3)}$.

### Desarrollo de los términos

**Término 1**

\begin{aligned}
\nabla[\textbf{u} \times (\nabla \times \pmb{\omega})] =
\partial_i u_j \partial_i \omega_j
+ u_j \partial^2_{ii} \omega_j
-\partial_i u_j \partial_j \omega_i
\end{aligned}

---

**Término 2**

\begin{aligned}
\textbf{u}(\nabla \times \nabla \times \pmb{\omega}) =
- u_i \partial^2_{jj} \omega_i
\end{aligned}

---

Luego,
sumando estos términos se obtiene:

\begin{aligned}
\pmb{\omega}(\nabla \times \pmb{\omega}) & =
\nabla[\textbf{u} \times (\nabla \times \pmb{\omega})]
+ \textbf{u}(\nabla \times \nabla \times \pmb{\omega}) \\
& = \partial_i u_j \partial_i \omega_j
+ u_j \partial^2_{ii} \omega_j
-\partial_i u_j \partial_j \omega_i
- u_i \partial^2_{jj} \omega_i \\
& = \partial_i u_j \partial_i \omega_j
- \partial_i u_j \partial_j \omega_i \\
\pmb{\omega}(\nabla \times \pmb{\omega})
+ \nabla [(\textbf{u} \nabla) \pmb\omega] 
& = \partial_i u_j \partial_i \omega_j
\tag{5}
\end{aligned}

De esta manera,
el término de $S$ que buscamos
puede ser introducido en la ecuación $(3)$
si conseguimos reescribir los términos disipativos
y hacer aparecer $\partial_i u_j \partial_i \omega_j$.
Para esto,
seguimos un proceso similar al que hicimos en el punto anterior
e intentamos introducir los vectores dentro del laplaciano.

\begin{aligned}
\nabla^2(\textbf{u} \cdot \pmb\omega) & =
u_i \partial^2_{jj} \omega_i
+ \omega_i \partial^2_{jj} u_i
+ 2 \partial_i u_j \partial_i \omega_j \\
& = \textbf{u} \nabla^2 \pmb\omega
+ \pmb\omega \nabla^2 \textbf{u}
+2 \pmb\omega (\nabla \times \pmb\omega)
+2 \nabla[(\textbf{u} \nabla)\pmb\omega]
\end{aligned}

Reemplazo los dos términos disipativos de la ecuación $(3)$
con los tres términos obtenidos en la última ecuación.

\begin{aligned}
\frac{\partial H}{\partial t} 
& = \nu \int_V \left(  \pmb{\omega} \nabla^2 \textbf{u}
+ \textbf{u} \nabla^2 \pmb{\omega} \right) \, dV \\
& =\nu \int_V \{ \nabla^2(\textbf{u} \cdot \pmb\omega) 
- 2 \pmb\omega (\nabla \times \pmb\omega)
- 2 \nabla[(\textbf{u} \nabla) \pmb\omega] \} \, dV \\
& = 0
- 2 \nu \int_V \pmb\omega (\nabla \times \pmb\omega) \, dV
- 0 \\
& \equiv - 2 \nu S
\end{aligned}

Donde puedo anular las integrales del primer
y último término utilizando el teorema de Gauss
y aplicando condiciones periódicas de contorno.


Otra manera de estimarlo es
calcular el "centro de masa"
de las regiones con helicidad positiva (y negativa).

```python
def center_of_mass(h, mask):
    """Devuelve un vector (x, y, z) con la posición del centro de masa.
    
    mask : np.ndarray[bool]
    """
    h = h.to_numpy()[mask]
    h /= h.sum()
    pos = np.where(mask)
    return xarray.DataArray([(h * p).sum() for p in pos])


def velocity(pos):
    """La norma del vector velocidad, calculado por diferencias finitas.
    
    || (x_1 - x_0) / dt ||
    """
    v = pos.diff("t")
    v = np.sqrt((v**2).sum(dim="dim_0"))
    return v


h = load_helicity(10)
pos = h.groupby("t").apply(lambda x: center_of_mass(x, x > 0))
neg = h.groupby("t").apply(lambda x: center_of_mass(x, x < 0))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

for label, x in {"h > 0": pos, "h < 0": neg}.items():
    x[:, 2].plot(label=label, ax=ax[0])
    velocity(x).plot(label=label, ax=ax[1])

ax[0].set(ylabel="Posición en z", xlabel="Tiempo")
ax[1].set(ylabel="Velocidad", xlabel="Tiempo")
fig.suptitle("Centro de helicidad h")
for a in ax: 
    ax[1].legend()
```

Después de un transitorio inicial,
la helicidad en $z$ parece moverse a velocidad constante.

El cambio de dirección (y velocidad) brusco en el final
se debe a que no se está teniendo en cuenta la periodicidad en el espacio.

```python
omega, pos, neg = [], [], []
for w, (p, n) in h_cm.items():
    omega.append(w)
    pos.append(velocity(p))
    neg.append(velocity(n))

pos = xarray.concat(pos, dim=xarray.IndexVariable("omega", omega))
neg = xarray.concat(neg, dim=xarray.IndexVariable("omega", omega))
```

## 3) Estratificación y energía total

Considere un fluído estratificado e incompresible de densidad media $\bar{\rho}$,
bajo la aproximación de Boussinesq.
El fluído satisfaces las siguientes ecuaciones,

\begin{aligned}
\frac{\partial \textbf{u}}{\partial t}
+ \textbf{u} \nabla \textbf{u} & =
- \nabla \left( \frac{p}{\rho} \right)
- N \theta \hat{z}
+ \nu \nabla^2 \textbf{u} \\
\frac{\partial \theta}{\partial t} 
+ \textbf{u} \nabla \theta & =
N u_z 
+ \kappa \nabla^2 \theta
\end{aligned}

Donde N es la frecuencia de Brunt-Väisälä uniforme,
$\theta$ es el potencial de las fluctuaciones de las temperaturas,
$\nu$ es la viscosidad cinemática y $\kappa$ es la difusividad térmica.

> a) Muestre que la energía cinética $E_c = \frac{\rho_0}{2} \int_V u^2 dV $ no se conserva en el caso ideal ($\nu = \kappa = 0$).
> Asuma condiciones de contorno periódicas.
> ¿Se le ocurren otras?
>
> b) Obtenga una ecuación de balance para la energía total (cinética más potencial):
\begin{align}
E = \frac{\rho_0}{2} \int_V u^2 dV + \frac{1}{2} \int_V \theta^2 dV
\end{align}
>
> c) ¿Qué cantidades controlan la tasa de disipación total? ¿Físicamente a qué corresponden?

### a)

Para obtener una ecuación de balance para la energia,
multiplicamos la ecuación $(5)$ por $\textbf{u}$:

Del primer termino,
se tiene:

$$ \textbf{u} \cdot \frac{\partial \textbf{u}}{\partial t}
= \frac{\partial (\textbf{u}^2 / 2)}{\partial t} $$

Del segundo término,
se tiene:
$$ \begin{aligned} \textbf{u} \cdot (\textbf{u} \cdot \nabla) \textbf{u}
&= (\textbf{u} \cdot \nabla) \; (\textbf{u}^2 / 2) \\
\text{por partes}
&= \nabla \; (\textbf{u} \; \textbf{u}^2 / 2)
- (\textbf{u}^2 / 2) \; \nabla \cdot \textbf{u} \\
\text{u incompresible}
&= \nabla \; (\textbf{u} \; \textbf{u}^2 / 2)
\end{aligned} $$

Para el tercer término,
$$ \textbf{u} \cdot \nabla \left(\frac{p}{\rho} \right)
= \nabla \left(\textbf{u} \frac{p}{\rho} \right)
$$
donde también se utiliza que $u$ es incompresible.

Para el cuarto término,
el que contiene a $\theta$,
al combinarlo con la otra ecuación,
se llega a:

$$ \begin{aligned}
\textbf{u} \cdot N \theta \hat{z}
&= N \theta u_z
\\ \\
&= \theta \; \frac{\partial \theta}{\partial t}
+ \theta \; \textbf{u} \nabla \theta
- \theta \; \kappa \nabla^2 \theta
\\ \\
&= \frac{\partial (\theta^2 / 2)}{\partial t}
+ \textbf{u} \nabla (\theta^2 / 2)
- \theta \; \kappa \nabla^2 \theta
\\ \\
&= \frac{\partial (\theta^2 / 2)}{\partial t}
+ \nabla (\textbf{u} \; \theta^2 / 2)
- \theta \; \kappa \nabla^2 \theta
\end{aligned} $$

donde también se utilizó la incompresibilidad del fluido
en el último paso.

Uniendo todo,
se obtiene la siguiente ecuación para $u^2 / 2$:

$$
\frac{\partial (\textbf{u}^2 / 2)}{\partial t}
+ \nabla \; (\textbf{u} \; \textbf{u}^2 / 2)
= - \frac{\partial (\theta^2 / 2)}{\partial t}
- \nabla (\textbf{u} \; \theta^2 / 2)
+ \theta \; \kappa \nabla^2 \theta
+ \textbf{u} \; \nu \nabla^2 \textbf{u}
- \nabla \left(\textbf{u} \frac{p}{\rho} \right)
$$

Multiplicando por $\rho$ e integrando en volumen,
obtenemos las variaciones de energía cinética $E_c$ y potencial $E_p$:

$$ \begin{aligned}
\int_{V} dV \rho \; \frac{\partial (\textbf{u}^2 / 2)}{\partial t}
= \frac{\partial}{\partial t} \int_{V} dV \rho \; (\textbf{u}^2 / 2)
= \frac{\partial E_c}{\partial t}
\\ \\
\int_{V} dV \rho \; \frac{\partial (\theta^2 / 2)}{\partial t}
= \frac{\partial E_p}{\partial t}
\end{aligned} $$

Para los términos con divergencias,
por el teorema de Stokes,
podemos cambiar la integral de la divergencia
por una integral sobre la superficie.
Esta integral se anula ya que tenemos condiciones de contorno periodicas.

$$ \begin{aligned}
&\int_{V} dV \rho \; \nabla \; (\textbf{u} \; \textbf{u}^2 / 2)
= \int_{S} dS \rho \; (\textbf{u} \; \textbf{u}^2 / 2)
= 0
\\
&\int_{V} dV \rho \; \nabla \;  (\textbf{u} \; \theta^2 / 2) = 0
\\
&\int_{V} dV \rho \; \nabla \left(\textbf{u} \frac{p}{\rho} \right) = 0
\end{aligned} $$

Considerando el caso ideal,
($\nu = \kappa = 0$),
llegamos a:

$$ \frac{\partial E_c}{\partial t} + \frac{\partial E_p}{\partial t} = 0 $$

## b)

En el caso no ideal,
tenemos:

$$ \frac{\partial E_c}{\partial t} + \frac{\partial E_p}{\partial t}
= \int_{V} dV \rho \; \theta \; \kappa \nabla^2 \theta
+ \int_{V} dV \rho \; \textbf{u} \; \nu \nabla^2 \textbf{u}
$$

Para el primer término,

$$ \int_{V} dV \rho \; \theta \; \kappa \nabla^2 \theta
= \int_{V} dV \rho \; \kappa \; \nabla^2 (\theta^2 / 2)
- \int_{V} dV \rho \; \kappa \; (\nabla \theta)^2
$$

donde el primer término se anula
por las condiciones de contorno periodicas.

Para el segundo término,

$$ \int_{V} dV \rho \; \textbf{u} \; \nu \nabla^2 \textbf{u}
= \int_{V} dV \rho \; \nabla^2 \textbf{u}
- \int_{V} dV \rho \; \textbf{u} \cdot ∇ × \textbf{u}
- \int_{V} dV \rho \; ∇ \Big( (\textbf{u} ⋅ ∇) \textbf{u} \Big)
$$

El primer y el tercer término se cancelan al integrar en superficie
por las condiciones de contorno periodicas.

El segundo término
podemos reescribirlo en términos de la vorticidad $ω = ∇ × \textbf{u}$,
donde su producto se conoce como la helicidad $h$.

Finalmente,
llegamos a la siguiente ecuación de balance para la energía:

$$ \frac{\partial E_c}{\partial t} + \frac{\partial E_p}{\partial t}
= \int_{V} dV \rho \; \kappa \; (\nabla \theta)^2
+ \int_{V} dV \rho \; \nu \; h
$$

### (b)
> Estudie la longitud de onda de la onda estacionaria
en función de la frecuencia de Brunt-Väisälä $N$,
variándola entre $1$ y $15$ (pasos de a $3$).

Para estimar la frecuencia de la onda,
buscamos sus picos a lo largo del eje $\hat{x}$,
calculamos la distancia entre picos,
y promediamos.
En particular,
elegimos un solo corte a $y=64$ y $z=64$,
para el tiempo final.

La frecuencia de Brunt-Väisälä
es la frecuencia angular de estas ondas:

```python
def calc_angular_frequency(n):
    T = load_temperature(n)
    Tx = T.isel(t=-1, y=64, z=64).to_numpy()
    ix, _ = find_peaks(Tx, height=Tx.max() / 10)
    period = T.x[ix].diff("x").mean()
    return 2 * np.pi / period

N = np.arange(1, 15, 3)[1:]
spatial_freq = np.array([calc_angular_frequency(n) for n in N])

plt.plot(N, spatial_freq, 'o--')
plt.xlabel("Frecuencia de Brunt-Väisälä N")
plt.ylabel("Frecuencia espacial")
plt.grid()
```
