## 1) Ley de Richardson

### a)
> Dos elementos de fluido en un flujo turbulento isótropo y homogéneo
están inicialmente a una distancia $λ$ dentro del rango inercial.
A partir de análisis dimensional,
muestre que esta distancia crece como,
> $$ \frac{dλ}{dt} ∼ (ελ)^{⅓} \tag{1}$$
> donde $ε$ es la tasa de disipación de energía por unidad de tiempo.

Como $U^2$ tiene unidades de energía (por unidad de masa),

$$ ε ∼ \frac{U²}{τ} = \frac{U³}{λ} $$

donde $U$, $τ$, y $λ$ son la velocidad, el tiempo y la longitud características.

Entonces,

$$ \frac{d λ}{dt} \sim U \sim (ε λ)^⅓ $$

## b)
> Deduzca entonces la Ley de Richardson para la difusión turbulenta:
el tiempo requerido para que la distancia entre los dos elementos de fluido sea
$λ₂$ (con $λ₂ ≫ λ₁$ ) es $τ ∼ (λ₂^2 / ε)^{⅓}$
(o, equivalentemente, que la distancia cuadrática que separa las partículas crece como el tiempo al cubo).

Se puede utilizar la misma relación que en el punto anterior,
entre el tiempo $\tau$,
la longitud y velocidad características $\lambda$ y $U$ respectivamente,
para despejar la ley de Richardos para la difusión turbulenta.

$$ τ
∼ \frac{λ}{U}
= \frac{λ}{(ε λ)^{⅓}}
= \left( \frac{λ²}{ε} \right)^{⅓}
$$

## 2)

### a)
> A partir de análisis dimensional,
muestre que para un flujo turbulento isótropo y homogéneo en ausencia de fuerzas externas
el balance de energía en el rango inercial se reduce a
> $$ \frac{∂E}{∂t} ∼ −ε \tag{2} $$

Multiplicando la ecuación de Navier-Stokes por $𝐮$,
tenemos:

$$ \frac{∂𝐮^2/2}{∂t} + 𝐮 · (𝐮 · ∇𝐮) = −𝐮 · ∇\frac{p}{ρ₀} + ν \, 𝐮 · ∇² 𝐮 $$

El término con $\nu$ lo podemos despreciar,
ya que estamos en el rango inercial.

Por otro lado,

$$ 𝐮 · ∇\frac{p}{ρ₀} = ∇ · \left(\frac{p}{ρ₀} 𝐮\right) - \frac{p}{ρ₀} ∇ · 𝐮 $$

El segundo término se anula,
al ser un fluido incompresible.
El primero se anula al considerar la integral en volumen,
si consideramos condiciones de contorno periodicas.

Entonces,

$$ \frac{∂E}{∂t} \sim - 𝐮 · (𝐮 · ∇𝐮) ∼ - \frac{U³}{L} ∼ - ε $$

### b)
> Asumiendo que la energía decae en forma auto–semejante,
$E(t) ∼ E₀ t^{−α}$
y que el fluido está contenido en un recipiente con longitud característica $L$,
muestre que $α = 2$.

Derivando con respecto al tiempo,
tenemos:

$$ \frac{∂E(t)}{∂t} = −α E₀ t^{−α−1} ∼ [E₀] τ^{−α−1}$$

Por otro lado,
según el punto anterior
$$ \frac{∂E(t)}{∂t} ∼ −ε $$

Entonces,
$$ [E₀] τ^{−α−1} ∼ \frac{∂E(t)}{∂t} ∼ −ε ∼ \frac{L²}{τ³} $$

Como $E_0$ no depende del tiempo,
$α = 2$.

## 3) Espectro de Kolmogorov y la ley de los 4/5

> Considere una magnitud escalar $θ$
que es advectada y difundida en forma pasiva
por un flujo incompresible turbulento isótropo y homogéneo con densidad uniforme $ρ₀$.
Las ecuaciones que describen la dinámica del sistema son
la ecuación de Navier-Stokes para el campo de velocidad $u$,
y la ecuación de advección para la concentración del escalar pasivo $θ$,
> $$ \frac{∂𝐮}{∂t} + 𝐮 · ∇𝐮 = −∇\frac{p}{ρ₀} + ν∇² 𝐮 \tag{1} $$
> $$ \frac{∂θ}{∂t} + 𝐮 · ∇θ = κ∇² θ \tag{2} $$
> donde κ es la difusividad del escalar pasivo.

### a)
> Asumiendo condiciones de contorno periódicas,
obtenga una ecuación de balance para la varianza del escalar pasivo,
> $$ Θ = \frac{1}{2} ⟨θ²⟩ = \frac{1}{2V} ∫ θ² dV $$
> y muestre que $Θ$ es una magnitud conservada cuando $κ = 0$.

Multiplicando la ecuación 2 por $θ$,
se obtiene:

$$ \frac{∂θ²}{∂t} + 𝐮 · ∇ θ² = 2κθ ∇² θ $$

Integrando en volumen,
se llega a:

$$ \begin{aligned}
2V \frac{dΘ}{dt}
&= ∫ \frac{∂θ²}{∂t} \;dV
\\ &= - ∫ 𝐮 · ∇ θ² \;dV + ∫ 2κ θ ∇² θ \;dV
\\ &= - ∫ [\underbrace{∇ (θ² 𝐮)  - θ² ∇·𝐮}_{\text{Se anulan}} ] \;dV + κ \left[ ∫ \underbrace{∇² θ²}_{\text{Se anula}} \;dV - ∫ (∇θ)² \;dV \right]
\\ &= - ∫ κ (∇θ)² \;dV
\end{aligned} $$

donde $ ∫ ∇(θ²𝐮) dV $ y $ ∫ ∇² θ² \, dV$ se pueden escribir como integrales de superficie,
que se anulan usando condiciones de contorno periodicas,
y el término $∇·𝐮$ se anula por incompresibilidad.

Entonces,
cuando $κ = 0$,
$\frac{dΘ}{dt} = 0$
y $Θ$ se conserva.

### b)
> Utilizando análisis dimensional,
muestre que en el rango inercial el flujo de la varianza del escalar pasivo,
> $ ε_θ ∼ \frac{dΘ}{dt} $
> satisface la siguiente relación:
> $$ ε_θ ∼ \frac{δθ²_ℓ δu_ℓ}{ℓ} $$
> donde $ℓ$ es una escala del rango inercial
y δ implica la función incremento.
Esta ley es equivalente a la ley de los 4/5 de Kolmogorov para el campo de velocidades.
Interprete este resultado en términos de una cascada turbulenta de la magnitud física.

Como

$$ \frac{dΘ}{dt} = \frac{1}{2V} ∫ κ θ ∇² θ \;dV $$

entonces,
$$ \epsilon_\theta
\sim \frac{dΘ}{dt}
\sim [\theta^2]
\underbrace{[L^2 / T]}_{\kappa}
\underbrace{[L^{-2}]}_{∇²}
\underbrace{[L^3]}_{dV}
\underbrace{[L^{-3}]}_{1/V}
= \frac{\theta^2}{T}
= \frac{\theta^2 U}{L}
$$

### c)
> Utilizando esta relación,
y asumiendo que el espectro de la energía cinética es el espectro de Kolmogorov,
> $ E(k) = C_K ε^{2/3} k^{−5/3} $,
> obtenga la predicción de Kolmogorov–Obukhov–Corrsin
para el espectro de la varianza del escalar pasivo,
> $ E_θ (k) = C_θ ε^{−1/3} ε_θ k^{−5/3} $.

Para el espectro del escalar pasivo,
tenemos que

$$ \frac{dE_\theta}{dk}
\sim \frac{E_\theta}{k}
\sim \frac{\theta^2}{k}
\sim \frac{\epsilon_\theta L}{k U}
$$

Como $U \sim (\epsilon L)^{1/3}$,

$$ \frac{dE_\theta}{dk}
\sim \frac{\epsilon_\theta L}{k (\epsilon L)^{1/3}}
\sim \frac{\epsilon_\theta}{k^{5/3} \epsilon^{1/3}}
$$

Luego,
esta expresión aproximada se puede transformar en una igualdad a través de una constante adecuada:

$$ \frac{dE_\theta}{dk}
= C_\theta \, \epsilon_\theta \, k^{-5/3} \epsilon^{-1/3}
$$

### (f)
> Utilizando la componente $x$ del campo de velocidad en un instante cercano al máximo de enstrofía,
calcule la función de estructura de segundo orden
> $$ S₂(r) = \Big\langle[uₓ(x + r x̂) − uₓ(x)]²\Big\rangle $$
> donde el valor medio es espacial sobre la coordenada $x$
(es decir, sobre todos los puntos del recinto).
Grafique $S₂(r)$ en escala log–log.
¿Obtiene el resultado esperado? ¿Por qué?

```python
vx = open_dataarray(
    "problema4/output",
    name="vx",
    dt=1.4,
    shape=(128, 128, 128),
    dtype=np.float32,
)
vx
```

```python
vx.isel(x=64, t=slice(0, 16, 4)).plot(col="t")
vx.isel(y=64, t=slice(0, 16, 4)).plot(col="t")
vx.isel(z=64, t=slice(0, 16, 4)).plot(col="t")
```

```python
import dask
from dask.diagnostics import ProgressBar


def second_order_structure(vx, r):
    diff = vx.roll(x=r) - vx
    return (diff**2).mean()


vx0 = vx.sel(t=5, method="nearest")
r = np.arange(vx0.x.size)
S2 = [second_order_structure(vx0, ri) for ri in r]

with ProgressBar():
    (S2,) = dask.compute(S2)
S2 = np.array(S2)
```

```python
def fit(x0, x1):
    ix = slice(x0, x1)
    p, cov = np.polyfit(np.log(r[ix]), np.log(S2[ix]), 1, cov=True)
    k, dk = p[0], cov[0, 0] ** 0.5
    plt.plot(
        r[ix],
        np.exp(np.polyval(p, np.log(r[ix]))),
        label=f"Exponente = {k:.3f} ± {dk:.3f}",
    )

plt.loglog(r, S2)
fit(20, 50)
fit(1, 5)
plt.legend()

plt.ylabel("$S_2(r)$")
plt.xlabel("r")
```

Obtenemos que,
en el rango inercial,
la función de escrutctura crece con un exponente $∼⅔$,
que es lo que propone Kolmogorov en su artículo de 1941
(Davidson P. A., *"Turbulence in Rotating, Stratified and Electrically Conducting Fields"*, p. 272).

Por otro lado,
para $r$ pequeño,
esperamos que la función de estructura siga un comportamiento del estilo $r^2$.
