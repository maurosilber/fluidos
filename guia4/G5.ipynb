## 1) Ley de Richardson

### a)
> Dos elementos de fluido en un flujo turbulento is√≥tropo y homog√©neo
est√°n inicialmente a una distancia $Œª$ dentro del rango inercial.
A partir de an√°lisis dimensional,
muestre que esta distancia crece como,
> $$ \frac{dŒª}{dt} ‚àº (ŒµŒª)^{‚Öì} \tag{1}$$
> donde $Œµ$ es la tasa de disipaci√≥n de energ√≠a por unidad de tiempo.

Como $U^2$ tiene unidades de energ√≠a (por unidad de masa),

$$ Œµ ‚àº \frac{U¬≤}{œÑ} = \frac{U¬≥}{Œª} $$

donde $U$, $œÑ$, y $Œª$ son la velocidad, el tiempo y la longitud caracter√≠sticas.

Entonces,

$$ \frac{d Œª}{dt} \sim U \sim (Œµ Œª)^‚Öì $$

## b)
> Deduzca entonces la Ley de Richardson para la difusi√≥n turbulenta:
el tiempo requerido para que la distancia entre los dos elementos de fluido sea
$Œª‚ÇÇ$ (con $Œª‚ÇÇ ‚â´ Œª‚ÇÅ$ ) es $œÑ ‚àº (Œª‚ÇÇ^2 / Œµ)^{‚Öì}$
(o, equivalentemente, que la distancia cuadr√°tica que separa las part√≠culas crece como el tiempo al cubo).

Se puede utilizar la misma relaci√≥n que en el punto anterior,
entre el tiempo $\tau$,
la longitud y velocidad caracter√≠sticas $\lambda$ y $U$ respectivamente,
para despejar la ley de Richardos para la difusi√≥n turbulenta.

$$ œÑ
‚àº \frac{Œª}{U}
= \frac{Œª}{(Œµ Œª)^{‚Öì}}
= \left( \frac{Œª¬≤}{Œµ} \right)^{‚Öì}
$$

### a)
> A partir de an√°lisis dimensional,
muestre que para un flujo turbulento is√≥tropo y homog√©neo en ausencia de fuerzas externas
el balance de energ√≠a en el rango inercial se reduce a
> $$ \frac{‚àÇE}{‚àÇt} ‚àº ‚àíŒµ \tag{2} $$

Multiplicando la ecuaci√≥n de Navier-Stokes por $ùêÆ$,
tenemos:

$$ \frac{‚àÇùêÆ^2/2}{‚àÇt} + ùêÆ ¬∑ (ùêÆ ¬∑ ‚àáùêÆ) = ‚àíùêÆ ¬∑ ‚àá\frac{p}{œÅ‚ÇÄ} + ŒΩ \, ùêÆ ¬∑ ‚àá¬≤‚ÄâùêÆ $$

El t√©rmino con $\nu$ lo podemos despreciar,
ya que estamos en el rango inercial.

Por otro lado,

$$ ùêÆ ¬∑ ‚àá\frac{p}{œÅ‚ÇÄ} = ‚àá ¬∑ \left(\frac{p}{œÅ‚ÇÄ} ùêÆ\right) - \frac{p}{œÅ‚ÇÄ} ‚àá ¬∑ ùêÆ $$

El segundo t√©rmino se anula,
al ser un fluido incompresible.
El primero se anula al considerar la integral en volumen,
si consideramos condiciones de contorno periodicas.

Entonces,

$$ \frac{‚àÇE}{‚àÇt} \sim - ùêÆ ¬∑ (ùêÆ ¬∑ ‚àáùêÆ) ‚àº - \frac{U¬≥}{L} ‚àº - Œµ $$

### b)

$$ \frac{‚àÇE(t)}{‚àÇt} ‚àº ‚àíŒµ $$

Entonces,
$$ [E‚ÇÄ] œÑ^{‚àíŒ±‚àí1} ‚àº \frac{‚àÇE(t)}{‚àÇt} ‚àº ‚àíŒµ ‚àº \frac{L¬≤}{œÑ¬≥} $$

Como $E_0$ no depende del tiempo,
$Œ± = 2$.

## 3) Espectro de Kolmogorov y la ley de los 4/5

> Considere una magnitud escalar $Œ∏$
que es advectada y difundida en forma pasiva
por un flujo incompresible turbulento is√≥tropo y homog√©neo con densidad uniforme $œÅ‚ÇÄ$.
Las ecuaciones que describen la din√°mica del sistema son
la ecuaci√≥n de Navier-Stokes para el campo de velocidad $u$,
y la ecuaci√≥n de advecci√≥n para la concentraci√≥n del escalar pasivo $Œ∏$,
> $$ \frac{‚àÇùêÆ}{‚àÇt} + ùêÆ ¬∑ ‚àáùêÆ = ‚àí‚àá\frac{p}{œÅ‚ÇÄ} + ŒΩ‚àá¬≤‚ÄâùêÆ \tag{1} $$
> $$ \frac{‚àÇŒ∏}{‚àÇt} + ùêÆ ¬∑ ‚àáŒ∏ = Œ∫‚àá¬≤‚ÄâŒ∏ \tag{2} $$
> donde Œ∫ es la difusividad del escalar pasivo.

### a)
> Asumiendo condiciones de contorno peri√≥dicas,
obtenga una ecuaci√≥n de balance para la varianza del escalar pasivo,
> $$ Œò = \frac{1}{2} ‚ü®Œ∏¬≤‚ü© = \frac{1}{2V} ‚à´ Œ∏¬≤‚ÄâdV $$
> y muestre que $Œò$ es una magnitud conservada cuando $Œ∫ = 0$.

Multiplicando la ecuaci√≥n 2 por $Œ∏$,
se obtiene:

$$ \frac{‚àÇŒ∏¬≤}{‚àÇt} + ùêÆ ¬∑ ‚àá Œ∏¬≤ = 2Œ∫Œ∏ ‚àá¬≤‚ÄâŒ∏ $$

Integrando en volumen,
se llega a:

$$ \begin{aligned}
2V \frac{dŒò}{dt}
&= ‚à´ \frac{‚àÇŒ∏¬≤}{‚àÇt} \;dV
\\ &= - ‚à´ ùêÆ ¬∑ ‚àá Œ∏¬≤ \;dV + ‚à´ 2Œ∫ Œ∏ ‚àá¬≤‚ÄâŒ∏ \;dV
\\ &= - ‚à´ [\underbrace{‚àá (Œ∏¬≤ ùêÆ)  - Œ∏¬≤ ‚àá¬∑ùêÆ}_{\text{Se anulan}} ] \;dV + Œ∫ \left[ ‚à´ \underbrace{‚àá¬≤‚ÄâŒ∏¬≤}_{\text{Se anula}} \;dV - ‚à´ (‚àáŒ∏)¬≤ \;dV \right]
\\ &= - ‚à´ Œ∫ (‚àáŒ∏)¬≤ \;dV
\end{aligned} $$

donde $ ‚à´ ‚àá(Œ∏¬≤ùêÆ) dV $ y $ ‚à´ ‚àá¬≤‚ÄâŒ∏¬≤ \, dV$ se pueden escribir como integrales de superficie,
que se anulan usando condiciones de contorno periodicas,
y el t√©rmino $‚àá¬∑ùêÆ$ se anula por incompresibilidad.

Entonces,
cuando $Œ∫ = 0$,
$\frac{dŒò}{dt} = 0$
y $Œò$ se conserva.

### b)
> Utilizando an√°lisis dimensional,
muestre que en el rango inercial el flujo de la varianza del escalar pasivo,
> $ Œµ_Œ∏ ‚àº \frac{dŒò}{dt} $
> satisface la siguiente relaci√≥n:
> $$ Œµ_Œ∏ ‚àº \frac{Œ¥Œ∏¬≤_‚Ñì Œ¥u_‚Ñì}{‚Ñì} $$
> donde $‚Ñì$ es una escala del rango inercial
y Œ¥ implica la funci√≥n incremento.
Esta ley es equivalente a la ley de los 4/5 de Kolmogorov para el campo de velocidades.
Interprete este resultado en t√©rminos de una cascada turbulenta de la magnitud f√≠sica.

Como

$$ \frac{dŒò}{dt} = \frac{1}{2V} ‚à´ Œ∫ Œ∏ ‚àá¬≤‚ÄâŒ∏ \;dV $$

entonces,
$$ \epsilon_\theta
\sim \frac{dŒò}{dt}
\sim [\theta^2]
\underbrace{[L^2 / T]}_{\kappa}
\underbrace{[L^{-2}]}_{‚àá¬≤}
\underbrace{[L^3]}_{dV}
\underbrace{[L^{-3}]}_{1/V}
= \frac{\theta^2}{T}
= \frac{\theta^2 U}{L}
$$

### c)
> Utilizando esta relaci√≥n,
y asumiendo que el espectro de la energ√≠a cin√©tica es el espectro de Kolmogorov,
> $ E(k) = C_K Œµ^{2/3} k^{‚àí5/3} $,
> obtenga la predicci√≥n de Kolmogorov‚ÄìObukhov‚ÄìCorrsin
para el espectro de la varianza del escalar pasivo,
> $ E_Œ∏ (k) = C_Œ∏ Œµ^{‚àí1/3} Œµ_Œ∏ k^{‚àí5/3} $.

Para el espectro del escalar pasivo,
tenemos que

$$ \frac{dE_\theta}{dk}
\sim \frac{E_\theta}{k}
\sim \frac{\theta^2}{k}
\sim \frac{\epsilon_\theta L}{k U}
$$

Como $U \sim (\epsilon L)^{1/3}$,

$$ \frac{dE_\theta}{dk}
\sim \frac{\epsilon_\theta L}{k (\epsilon L)^{1/3}}
\sim \frac{\epsilon_\theta}{k^{5/3} \epsilon^{1/3}}
$$

Luego,
esta expresi√≥n aproximada se puede transformar en una igualdad a trav√©s de una constante adecuada:

$$ \frac{dE_\theta}{dk}
= C_\theta \, \epsilon_\theta \, k^{-5/3} \epsilon^{-1/3}
$$

> Utilizando la componente $x$ del campo de velocidad en un instante cercano al m√°ximo de enstrof√≠a,
calcule la funci√≥n de estructura de segundo orden
> $$ S‚ÇÇ(r) = \Big\langle[u‚Çì(x + r xÃÇ) ‚àí u‚Çì(x)]¬≤\Big\rangle $$
> donde el valor medio es espacial sobre la coordenada $x$
(es decir, sobre todos los puntos del recinto).
Grafique $S‚ÇÇ(r)$ en escala log‚Äìlog.
¬øObtiene el resultado esperado? ¬øPor qu√©?

```python
vx = open_dataarray(
    "problema4/output",
    name="vx",
    dt=1.4,
    shape=(128, 128, 128),
    dtype=np.float32,
)
vx
```

```python
vx.isel(x=64, t=slice(0, 16, 4)).plot(col="t")
vx.isel(y=64, t=slice(0, 16, 4)).plot(col="t")
vx.isel(z=64, t=slice(0, 16, 4)).plot(col="t")
```

```python
import dask
from dask.diagnostics import ProgressBar


def second_order_structure(vx, r):
    diff = vx.roll(x=r) - vx
    return (diff**2).mean()


vx0 = vx.sel(t=5, method="nearest")
r = np.arange(vx0.x.size)
S2 = [second_order_structure(vx0, ri) for ri in r]

with ProgressBar():
    (S2,) = dask.compute(S2)
S2 = np.array(S2)
```

```python
def fit(x0, x1):
    ix = slice(x0, x1)
    p, cov = np.polyfit(np.log(r[ix]), np.log(S2[ix]), 1, cov=True)
    k, dk = p[0], cov[0, 0] ** 0.5
    plt.plot(
        r[ix],
        np.exp(np.polyval(p, np.log(r[ix]))),
        label=f"Exponente = {k:.3f} ¬± {dk:.3f}",
    )

plt.loglog(r, S2)
fit(20, 50)
fit(1, 5)
plt.legend()

plt.ylabel("$S_2(r)$")
plt.xlabel("r")
```

Obtenemos que,
en el rango inercial,
la funci√≥n de escrutctura crece con un exponente $‚àº‚Öî$,
que es lo que propone Kolmogorov en su art√≠culo de 1941
(Davidson P. A., *"Turbulence in Rotating, Stratified and Electrically Conducting Fields"*, p. 272).

Por otro lado,
para $r$ peque√±o,
esperamos que la funci√≥n de estructura siga un comportamiento del estilo $r^2$.
